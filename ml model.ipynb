```python
# splitting the dataset into independent(x) and dependent dataset variable vector(y)
x = dataset.iloc[: , 3:13].values
y = dataset.iloc[: , -1].values
x
```

```python
# encoding the gender column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[: , 2] = le.fit_transform(x[: , 2])
x


# 0 ---> female , 1---> male
```

```python
# encoding the Geography column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
cm = ColumnTransformer(transformers=[("Geography" , OneHotEncoder() , [1])] , remainder= "passthrough")
x = np.array(cm.fit_transform(x))


# removing dummy variable trap

x = x[: , 1:]
x
# 0.0, 0.0 ---> France
# 0.0, 1.0 ---> Spain
# 1.0, 0.0 ---> germany
```

```python
# splitting the dataset into training and test dataset
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = .20 , random_state = 0)
print(x_train)
print(x_test)
```

```python
"""feature scalling the x_train and x_test"""
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
print(x_train)
```

```python
# applying kernel PCA ---> dimention reduction 

#from sklearn.decomposition import KernelPCA 
#kpca = KernelPCA(n_components = 2, kernel = 'rbf')
#x_train = kpca.fit_transform(x_train , y_train)
#x_test = kpca.transform(x_test)

```

```python
# applying K fold cross validation on the training dataset
#from sklearn.model_selection import cross_val_score
#cross_val_score()
```

```python
# appling ML model om the training dataset
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators= 500 , n_jobs= -1 , random_state= 0)
classifier.fit(x_train , y_train)

```

```python
# predicting new values from the ml model 
y_preg = classifier.predict(x_test)
```

```python
# making a confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_preg)
cm

# {1519 + 212} ---> correct prediction
# {193 + 76} ---> incorrect prediction
```

```python
# accuracy on 1 fold

accuracy = (1519 + 212) / (1519 + 212 + 76 + 193)
print(accuracy)
```

```python
# implementing K- cross validation 

from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(estimator= classifier , X = x_train , y = y_train , cv = 10 , n_jobs= -1)
score = accuracy.mean()
std = accuracy.std()

print("accuracy of the model is " , (score*100) , "%" )
print("standard deviation of the model is " , std )


# accuracy on 10 folds
```
